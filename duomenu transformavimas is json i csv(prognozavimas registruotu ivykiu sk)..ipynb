```python
import glob
import json
import pandas as pd
import re


data = {}
for file_path in glob.glob('prir_*.json'):
    with open(file_path, 'r', encoding='utf-8') as f:
        data[file_path] = json.load(f)
```

```python
dfs = []
for file_path in sorted(glob.glob('prir_*.json')):
    date_match = re.search(r'prir_(\d{4}_\d{2}_\d{2})', file_path)
    date = date_match.group(1).replace('_', '-') if date_match else 'unknown'

    with open(file_path, 'r', encoding='utf-8') as f:
        df = pd.json_normalize(json.load(f))  
        df['report_date'] = date
        dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df.head())
```

```python
display(combined_df)
```

```python
combined_df['registravimoData'] = pd.to_datetime(combined_df['registravimoData'], errors='coerce')

combined_df.to_csv('prir_combined_crime_data.csv', index=False, encoding='utf-8')
```
